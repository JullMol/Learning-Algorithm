```python
import pandas as pd

fitur_dashboard = [
    "indeks_pembangunan_manusia_(ipm)",
    "pengeluaran_per_kapita_riil_disesuaikan_(ribu_rupiah)",
    "tingkat_pengangguran_terbuka_(tpt)_-_agustus",
    "gini_ratio",
    "persentase_rumah_tangga_dengan_akses_air_bersih",
    "jumlah_penerima_bansos"
]

df_filtered = df[df["tahun"].between(2020, 2024)].copy()


def normalize_gini(x):
    if x > 100:
        return x / 1000
    elif x > 1:
        return x / 100
    else:
        return x

# Terapkan normalisasi
df_filtered["gini_ratio"] = df_filtered["gini_ratio"].apply(normalize_gini)

# Hitung rata-rata seluruh kabupaten/kota (gabungan Jatim)
jatim_rata2 = df_filtered[fitur_dashboard].mean().reset_index()
jatim_rata2.columns = ["fitur", "rata2_2020_2024"]

print(jatim_rata2)
```

```python
# Urutkan data berdasarkan 'tahun' dan 'kabupaten/kota'
df.sort_values(by=['kabupaten/kota', 'tahun'], inplace=True)

df['laju_pertumbuhan_ipm'] = df.groupby('kabupaten/kota')['indeks_pembangunan_manusia_(ipm)'].pct_change()
df['laju_pertumbuhan_pengeluaran_riil'] = df.groupby('kabupaten/kota')['pengeluaran_per_kapita_riil_disesuaikan_(ribu_rupiah)'].pct_change()
df['laju_pertumbuhan_tpt'] = df.groupby('kabupaten/kota')['tingkat_pengangguran_terbuka_(tpt)_-_agustus'].pct_change()
```

```python
df['kemiskinan_tahun_sebelumnya'] = df.groupby('kabupaten/kota')['persentase_penduduk_miskin_(persen)'].shift(1)
df.dropna(inplace=True)
```

```python
df["jumlah_industri_total"] = (
    df["jmlh_industri_q1"] +
    df["jmlh_industri_q2"] +
    df["jmlh_industri_q3"] +
    df["jmlh_industri_q4"]
)
```

```python
import pandas as pd
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

fitur = [
    'indeks_pembangunan_manusia_(ipm)',
    'tingkat_pengangguran_terbuka_(tpt)_-_agustus',
    'pengeluaran_per_kapita_riil_disesuaikan_(ribu_rupiah)',
    'gini_ratio',
    'jumlah_penerima_bansos',
    'kemiskinan_tahun_sebelumnya',
    'harapan_lama_sekolah_(tahun)',
    'status_balita_pendek',
    'produksi_padi_(ton)',
    'jumlah_industri_total',
    'laju_pertumbuhan_penduduk_per_tahun_(persen)',
    'kepadatan_penduduk_per_km_persegi_(km²)'
]

target = 'persentase_penduduk_miskin_(persen)'

X = df[fitur]
y = df[target]

X_train = df[df['tahun'] < 2024][fitur]
y_train = df[df['tahun'] < 2024][target]

X_test = df[df['tahun'] == 2024][fitur]
y_test = df[df['tahun'] == 2024][target]

print(f"Jumlah data training: {len(X_train)} baris")
print(f"Jumlah data testing: {len(X_test)} baris")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

model = xgb.XGBRegressor(
    n_estimators=1000,
    learning_rate=0.05,
    max_depth=5,
    subsample=0.8,
    colsample_bytree=0.8,
    objective='reg:squarederror',
    random_state=42,
    n_jobs=-1
)

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print(f"\nEvaluasi Model pada Data Uji Tahun 2024:")
print(f"Mean Squared Error (MSE): {mse:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
print(f"R-squared (R2): {r2:.4f}")
```

```python
df_historical_growth = df[(df['tahun'] >= 2020) & (df['tahun'] <= 2024)]
df_historical_growth = df_historical_growth.fillna(0)  # atau metode lain sesuai kebutuhan

numeric_cols = df_historical_growth.select_dtypes(include=['int64','float64']).columns
avg_growth = df_historical_growth.groupby('tahun')[numeric_cols].mean().pct_change().mean()

fitur = [
    'indeks_pembangunan_manusia_(ipm)',
    'tingkat_pengangguran_terbuka_(tpt)_-_agustus',
    'pengeluaran_per_kapita_riil_disesuaikan_(ribu_rupiah)',
    'gini_ratio',
    'jumlah_penerima_bansos',
    'kemiskinan_tahun_sebelumnya',
    'harapan_lama_sekolah_(tahun)',
    'status_balita_pendek',
    'produksi_padi_(ton)',
    'jumlah_industri_total',
    'laju_pertumbuhan_penduduk_per_tahun_(persen)',
    'kepadatan_penduduk_per_km_persegi_(km²)'
]

tahun_proyeksi = [2025, 2026, 2027]

df_base = df[df['tahun'] == 2024].copy()
df_forecast_list = []

for t in tahun_proyeksi:
    df_next = df_base.copy()
    df_next['tahun'] = t
    for col in fitur:
        df_next[col] *= (1 + avg_growth[col])
    df_forecast_list.append(df_next)
    df_base = df_next

df_forecast = pd.concat(df_forecast_list, ignore_index=True)

forecast_kemiskinan = model.predict(df_forecast[fitur])
df_forecast['kemiskinan_prediksi'] = forecast_kemiskinan

print(df_forecast[['tahun'] + fitur + ['kemiskinan_prediksi']])
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Deteksi kabupaten/kota dengan paradoks
paradoks_list = []
for kab in df['kabupaten/kota'].unique():
    # Data historis kabupaten
    hist = df[df['kabupaten/kota'] == kab][['tahun', 'persentase_penduduk_miskin_(persen)']]
    pred = df_forecast[df_forecast['kabupaten/kota'] == kab][['tahun', 'kemiskinan_prediksi']]

    if len(hist) > 1 and len(pred) > 1:
        # slope historis
        X_hist = hist['tahun'].values.reshape(-1, 1)
        y_hist = hist['persentase_penduduk_miskin_(persen)'].values
        slope_hist = LinearRegression().fit(X_hist, y_hist).coef_[0]

        # slope prediksi
        X_pred = pred['tahun'].values.reshape(-1, 1)
        y_pred = pred['kemiskinan_prediksi'].values
        slope_pred = LinearRegression().fit(X_pred, y_pred).coef_[0]

        # cek paradoks (berlawanan arah)
        if slope_hist * slope_pred < 0:
            paradoks_list.append(kab)

print("Kabupaten/Kota dengan tren paradoks:", paradoks_list)

# Gabungkan: top3, bottom3, dan paradoks
selected_kab = list(set(top3_kab + bottom3_kab + paradoks_list))

# Dataset historis + prediksi untuk gabungan
df_hist_sel = df[df['kabupaten/kota'].isin(selected_kab)][
    ['tahun', 'kabupaten/kota', 'persentase_penduduk_miskin_(persen)']
].rename(columns={'persentase_penduduk_miskin_(persen)': 'persentase_kemiskinan'})

df_pred_sel = df_forecast[df_forecast['kabupaten/kota'].isin(selected_kab)][
    ['tahun', 'kabupaten/kota', 'kemiskinan_prediksi']
].rename(columns={'kemiskinan_prediksi': 'persentase_kemiskinan'})

data_gabungan_sel = pd.concat([df_hist_sel, df_pred_sel], ignore_index=True)

# Plot grafik
plt.figure(figsize=(12, 8))
sns.lineplot(
    data=data_gabungan_sel,
    x='tahun',
    y='persentase_kemiskinan',
    hue='kabupaten/kota',
    style='kabupaten/kota',  # bedakan garis
    marker='o'
)

plt.axvline(x=df['tahun'].max() + 0.5, color='gray', linestyle='--', label='Perbatasan Prediksi')

# Highlight khusus untuk paradoks
for kab in paradoks_list:
    plt.text(
        x=data_gabungan_sel['tahun'].max(),
        y=data_gabungan_sel[data_gabungan_sel['kabupaten/kota'] == kab]['persentase_kemiskinan'].iloc[-1],
        s=f"← {kab} (Paradoks)",
        fontsize=10,
        color="red"
    )

plt.title('Tren Kemiskinan: Top 3, Bottom 3 & Paradoks di Jawa Timur', fontsize=16)
plt.xlabel('Tahun', fontsize=12)
plt.ylabel('Persentase Penduduk Miskin (%)', fontsize=12)
plt.legend(title="Kabupaten/Kota", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```
